### SELECTING WANTED COLUMNS

```python
data = pd.read_excel('seekingalpha_attachment.xlsx',sheet_name=1,header=5,skipfooter=16)
data = data[['Name','Symbol','Sector','Industry','Yrs','Price','Yield','Year','1-yr','3-yr','5-yr','10-yr','P/E','ROE','($Mil)','Equity','Survived','ROA']]
data
```

### CHANGING TYPES FROM STRING TO FLOAT

```python
str_to_float_columns = ['Yrs','Price','Yield','Year','1-yr','3-yr','5-yr','10-yr','P/E','ROE','($Mil)','Equity','Survived','ROA']
for column in str_to_float_columns:
    data[column] = pd.to_numeric(data[column])
data.info()
```

### IMPOSING RESTRICTIONS IN ORDER TO SELECT LISTED COMPANIES THAT I AM INTERESTED IN

```python
data.loc[(data['Yield'] > 2) & (data['Yield'] < 6) & (data['P/E'] < 20) & (data['ROE'] > 10) & (data['($Mil)'] > 10000) & 
         (data['Equity'] < 1)]

data.loc[data['ROE'] > 10]
data.loc[data['($Mil)'] > 10000]
data.loc[data['Equity'] < 1]
data
```
